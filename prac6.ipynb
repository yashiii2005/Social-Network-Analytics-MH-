```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
```

**SIR AND SIS TREND IN Erdos and Renyi Model**

```python
N=1000
p=0.1
G=nx.erdos_renyi_graph(N,p)
#G=nx.barabasi_albert_graph(1000,3)
#G=nx.watts_strogatz_graph(1000,6,0.1)
```

```python
beta=0.3
gamma=0.1
timesteps=50
initial_infected_ratio=0.01
```

```python
def simulate_SIS(G,beta,gamma,timesteps,initial_infected_ratio):
  nodes=list(G.nodes())
  infected=set(random.sample(nodes,int(initial_infected_ratio*len(nodes))))
  infected_counts=[len(infected)]
  for t in range(timesteps):
    new_infected=set(infected)

    for node in list(infected):
      if random.random()<gamma:
        new_infected.remove(node)
    ##Infection
    for node in list(G.nodes()):
      if node not in infected:
        for neighbor in G.neighbors(node):
          if neighbor in infected and random.random()<beta:
            new_infected.add(node)
            break
    infected=new_infected
    infected_counts.append(len(infected))
  return infected_counts
```

```python
def simulate_SIR(G,beta,gamma,timesteps,initial_infected_ratio):
  nodes=list(G.nodes())
  infected=set(random.sample(nodes,int(initial_infected_ratio*len(nodes))))
  recovered=set()
  infected_counts=[len(infected)]

  for t in range(timesteps):
    new_infected=set(infected)

    for node in G.nodes():
      if node not in infected and node not in recovered:
        for neighbor in G.neighbors(node):
          if neighbor in infected and random.random()<beta:
            new_infected.add(node)
            break
        for node in list(infected):
          if random.random()<gamma:
            recovered.add(node)

        infected=new_infected
        infected_counts.append(len(infected))
  return infected_counts
```

```python
sis_trend=simulate_SIS(G,beta,gamma,timesteps,initial_infected_ratio)
```

```python
plt.figure(figsize=(9,6))
plt.plot(sis_trend,label='SIS Model',linewidth=2)
plt.title('Different Trends in SIS Model (Erdos-Renyi Model)')
plt.xlabel('Time Steps')
plt.ylabel('Number of Infected Nodes')
plt.legend()
plt.grid(True)
plt.show()

```

```python
sir_trend=simulate_SIR(G,beta,gamma,timesteps,initial_infected_ratio)
```

```python
plt.figure(figsize=(9,6))
plt.plot(sir_trend,label='SIR Model',linewidth=2,linestyle="--")
plt.title('Different Trends in SIR Model (Erdos-Renyi Model)')
plt.xlabel('Time Steps')
plt.ylabel('Number of Infected')
plt.legend()
plt.grid(True)
plt.show()
```
