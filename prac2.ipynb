```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#weighted directed network with edge width and node size proportional to edge weight and degree
```

```python
wdirectedg = nx.DiGraph()
edges=([(1,2,3),(1,4,6),(2,3,2),(3,4,1),(4,5,5),(4,3,5),(5,1,4),(5,6,6),(6,6,1),(6,2,1)])
wdirectedg.add_weighted_edges_from(edges)

tdegree=dict(wdirectedg.degree())
print("total degree (in +out) : ", tdegree)
```

```python
node_size = []
for n in wdirectedg.nodes():
    scaled_degree = tdegree[n]*500
    node_size.append(scaled_degree)
print("node size : ",node_size)
```

```python
edge_weight=[]
for e in wdirectedg.edges():
    edge_weight.append(wdirectedg[e[0]][e[1]]['weight'])
print("edge weight : ",edge_weight)
```

```python
plt.figure(figsize=(6,6))
pos=nx.spring_layout(wdirectedg) 
nx.draw(wdirectedg,pos, with_labels=True, node_color='grey', node_size=node_size, width=edge_weight)
labels=nx.get_edge_attributes(wdirectedg,'weight')            #get edge weight
nx.draw_networkx_edge_labels(wdirectedg, pos, edge_labels=labels)
plt.title("weighted directed network with edge width and node size proportional to edge weight and degree")
plt.show()
```

```python
#weighted directed network with edge width and node size proportional to edge weight and inward degree
```

```python
wdirectedg = nx.DiGraph()
edges=([(1,2,3),(1,4,6),(2,3,2),(3,4,1),(4,5,5),(4,3,5),(5,1,4),(5,6,6),(6,6,1),(6,2,1)])
wdirectedg.add_weighted_edges_from(edges)

indegree=dict(wdirectedg.in_degree())
print("total inward degree : ", indegree)

node_size = []
for n in wdirectedg.nodes():
    scaled_degree = indegree[n]*500
    node_size.append(scaled_degree)
print("node size : ",node_size)

edge_weight=[]
for e in wdirectedg.edges():
    edge_weight.append(wdirectedg[e[0]][e[1]]['weight'])
print("edge weight : ",edge_weight)

plt.figure(figsize=(6,6))
pos=nx.spring_layout(wdirectedg) 
nx.draw(wdirectedg,pos, with_labels=True, node_color='grey', node_size=node_size, width=edge_weight)
labels=nx.get_edge_attributes(wdirectedg,'weight')            #get edge weight
nx.draw_networkx_edge_labels(wdirectedg, pos, edge_labels=labels)
plt.title("weighted directed network with edge width and node size proportional to edge weight and inward degree")
plt.show()
```
