```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# RANDOM NETWORK MODEL
```

```python
#RNM
n=7
p=.77
```

```python
G_rn=nx.erdos_renyi_graph(n,p)
nx.draw(G_rn,with_labels=True)
nx.draw(G_rn, with_labels=True,node_color="grey",node_size=500, font_weight='bold')
plt.show()
```

```python
print("no. of nodes : ", G_rn.number_of_nodes())
print("no. of edges : ", G_rn.number_of_edges())
print("avg. degree : ", sum(dict(G_rn.degree()).values())/ G_rn.number_of_nodes())
print("node clustering coefficient : ")
for node in G_rn.nodes():
    print(node, " ; ", nx.clustering(G_rn,node))
print("avg. clustering coefficient : ", nx.average_clustering(G_rn))
bc=nx.betweenness_centrality(G_rn)
```
