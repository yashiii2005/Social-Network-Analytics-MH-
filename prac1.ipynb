```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 1. SIMPLE GRAPH (UNDIRECTED & UNWEIGHTED )
```

```python
simpleg=nx.Graph()
simpleg.add_edges_from([(1,2),(2,3),(3,4),(4,5),(5,1)])

plt.figure(figsize=(4,4))
nx.draw(simpleg, with_labels=True, node_color='pink', node_size=550)
plt.title("simple (undirected and unweighted) graph")
plt.show()
```

```python
# DIRECTED GRAPH (UNWEIGHTED)
```

```python
directedg = nx.DiGraph()
directedg.add_edges_from([(1,2),(1,4),(2,3),(3,4),(4,5),(4,3),(5,1)])


plt.figure(figsize=(4,4))
nx.draw(directedg, with_labels=True, node_color='pink', node_size=550,arrows=True)
plt.title("directed (unweighted) graph")
plt.show()
```

```python
# DIRECTED WEIGHTED GRAPH
```

```python
wdirectedg = nx.DiGraph()
edges=([(1,2,3),(1,4,6),(2,3,2),(3,4,1),(4,5,5),(4,3,5),(5,1,4)])
wdirectedg.add_weighted_edges_from(edges)

pos=nx.spring_layout(wdirectedg)        #position nodes
plt.figure(figsize=(4,4))
nx.draw(wdirectedg,pos, with_labels=True, node_color='pink', node_size=550,arrows=True)
labels=nx.get_edge_attributes(wdirectedg,'weight')            #get edge weight
nx.draw_networkx_edge_labels(wdirectedg, pos, edge_labels=labels)
plt.title("directed weighted graph")
plt.show()
```
